```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline

df = pd.read_csv("./data/pima-data.csv")
df.shape

df.head(5)
df.tail(5)
df.isnull().values.any()

def plot_corr(df, size=14):
    corr = df.corr() # correlation function
    fig, ax = plt.subplots(figsize=(size, size))
    ax.matshow(corr)
    plt.xticks(range(len(corr.columns)), corr.columns) # x-axis ticks
    plt.yticks(range(len(corr.columns)), corr.columns) # y-axis ticks


plot_corr(df)
df.corr()
```

```python
plot_corr(df)
```

```python
dia_map = {True: 1, False: 0}
df['diabetes'] = df['diabetes'].map(dia_map)
```

```python
df.head(5)
```

```python
num_true = len(df.loc[df['diabetes'] == True])
num_false = len(df.loc[df['diabetes'] == False])

print("Number of True cases: {0} ({1:2.2f}%)".format(num_true, (num_true/ (num_true + num_false)) *100))
print("Number of True cases: {0} ({1:2.2f}%)".format(num_false, (num_true/ (num_true + num_false)) *100))      
```
